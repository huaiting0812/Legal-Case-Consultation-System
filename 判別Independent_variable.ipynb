```python
import pandas as pd
import jieba as jb
import numpy as np
```

+ 將code、關鍵字檔案（keyword.txt）、處理雜字的檔案（stopword.txt）放在同一個位址或資料夾中。
+ stopword_list就變成雜字的list，而`jb.load_userdict('keyword.txt')`則是將字定義字典加入到jieba分詞的字典內。

```python
jb.load_userdict('keyword.txt')
stopword = open('stopword.txt')
stopword_f = open('stopword.txt',encoding='utf-8')
tem = stopword_f.read()
tem_list = tem.split('\n')
tem_list[-3] = '\u3000'
tem_list[-2] = '\xa0'
tem_list[-5] = '\n'
tem_list.append(' ')
stopword_list = tem_list
```

```python
#file為要放入的判決書資料
file = ''
temp_data = pd.read_csv(file)
text = temp_data.full_text
```

```python
crime_df = pd.DataFrame(columns = ['有無酒駕','有無駕照','有無證據','有無肇事逃逸',
                                   '肇事車輛類型','有無累犯','有無自首','有無服用非法藥物',
                                   '有無超速'],
                        index = range(1,1+len(text)))
```

```python
#判斷有無並給值

for i in range(1,len(text)):
    words = jb.cut(text[i-1])
    words_list = []
    for x in words:
        if x not in stopword_list:
            words_list.append(x)
    
    for x in words_list:
        if x == '酒駕' or x == '酒後駕車' or x == '犯駕駛動力交通工具':
            crime_df.有無酒駕[i] = 1
            break
        else:
            crime_df.有無酒駕[i] = 0
            
    for x in words_list:
        if x == '無照' or x == '無駕駛執照':
            crime_df.有無駕照[i] = 0
            break
        else:
            crime_df.有無駕照[i] = 1
            
    for x in words_list:
        if x == '證據' and x != '證據不足':
            crime_df.有無證據[i] = 1
            break
        else:
            crime_df.有無證據[i] = 0
            
    for x in words_list:
        if x == '肇事逃逸'or x == '肇事後逃逸' :
            crime_df.有無肇事逃逸[i] = 1
            break
        else:
            crime_df.有無肇事逃逸[i] = 0
            
    change_index = 0
    for x in words_list:
        if (x == '自用小客貨車' or x == '小客貨車' or x == '自用小貨車') and crime_df.肇事車輛類型[i]!= 1:
            crime_df.肇事車輛類型[i] = 1
            change_index += 1
            if change_index >= 2:
                crime_df.肇事車輛類型[i] = 'None'
                break
        elif (x == '自用小客車' or x == '自小客車') and crime_df.肇事車輛類型[i]!= 2:
            crime_df.肇事車輛類型[i] = 2
            change_index += 1
            if change_index >= 2:
                crime_df.肇事車輛類型[i] = 'None'
                break
        
        elif (x == '普通重型機車' or x == '普通重型機') and crime_df.肇事車輛類型[i]!= 3:
            crime_df.肇事車輛類型[i] = 3
            change_index += 1
            if change_index >= 2:
                crime_df.肇事車輛類型[i] = 'None'
                break
        
        elif (x == '重型機車') and crime_df.肇事車輛類型[i]!= 4:
            crime_df.肇事車輛類型[i] = 4
            change_index += 1
            if change_index >= 2:
                crime_df.肇事車輛類型[i] = 'None'
                break
            
        elif (x == '輕型機車') and crime_df.肇事車輛類型[i]!= 5:
            crime_df.肇事車輛類型[i] = 5
            change_index += 1
            if change_index >= 2:
                crime_df.肇事車輛類型[i] = 'None'
                break
    
    for x in words_list:
        if x == '累犯':
            crime_df.有無累犯[i] = 1
            break
        else:
            crime_df.有無累犯[i] = 0
            
    for x in words_list:
        if x == '自首' :
            crime_df.有無自首[i] = 1
            break
        else:
            crime_df.有無自首[i] = 0
            
    for x in words_list:
        if x == '施用第一級毒品' or x == '施用第二級毒品' or x == '施用第三級毒品':
            crime_df.有無服用非法藥物[i] = 1
            break
        else:
            crime_df.有無服用非法藥物[i] = 0
            
    for x in words_list:
        if x == '超速' :
            crime_df.有無超速[i] = 1
            break
        else:
            crime_df.有無超速[i] = 0
```
